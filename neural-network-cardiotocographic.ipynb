# **Pre-Processing of the data**

# **Normalization of the variables**

```r
cardio$LB = (cardio$LB-min(cardio$LB))/(max(cardio$LB)-min(cardio$LB))

cardio$AC = (cardio$AC-min(cardio$AC))/(max(cardio$AC)-min(cardio$AC))

cardio$FM = (cardio$FM-min(cardio$FM))/(max(cardio$FM)-min(cardio$FM))

cardio$UC = (cardio$UC-min(cardio$UC))/(max(cardio$UC)-min(cardio$UC))

cardio$DL = (cardio$DL-min(cardio$DL))/(max(cardio$DL)-min(cardio$DL))

cardio$DS = (cardio$DS-min(cardio$DS))/(max(cardio$DS)-min(cardio$DS))

cardio$DP = (cardio$DP-min(cardio$DP))/(max(cardio$DP)-min(cardio$DP))

cardio$ASTV = (cardio$ASTV-min(cardio$ASTV))/(max(cardio$ASTV)-min(cardio$ASTV))

cardio$MSTV = (cardio$MSTV-min(cardio$MSTV))/(max(cardio$MSTV)-min(cardio$MSTV))

cardio$ALTV = (cardio$ALTV-min(cardio$ALTV))/(max(cardio$ALTV)-min(cardio$ALTV))

cardio$MLTV = (cardio$MLTV-min(cardio$MLTV))/(max(cardio$MLTV)-min(cardio$MLTV))

cardio$Width = (cardio$Width-min(cardio$Width))/(max(cardio$Width)-min(cardio$Width))

cardio$Min = (cardio$Min-min(cardio$Min))/(max(cardio$Min)-min(cardio$Min))

cardio$Max = (cardio$Max-min(cardio$Max))/(max(cardio$Max)-min(cardio$Max))

cardio$Nmax = (cardio$Nmax-min(cardio$Nmax))/(max(cardio$Nmax)-min(cardio$Nmax))

cardio$Nzeros = (cardio$Nzeros-min(cardio$Nzeros))/(max(cardio$Nzeros)-min(cardio$Nzeros))

cardio$Mode = (cardio$Mode-min(cardio$Mode))/(max(cardio$Mode)-min(cardio$Mode))

cardio$Mean = (cardio$Mean-min(cardio$Mean))/(max(cardio$Mean)-min(cardio$Mean))

cardio$Median = (cardio$Median-min(cardio$Median))/(max(cardio$Median)-min(cardio$Median))

cardio$Variance = (cardio$Variance-min(cardio$Variance))/(max(cardio$Variance)-min(cardio$Variance))

cardio$Tendency = (cardio$Tendency-min(cardio$Tendency))/(max(cardio$Tendency)-min(cardio$Tendency))
```

```r
# Structure of the dataset
str(cardio)
```

# **Data Partition**

```r
# Setting seed 
set.seed(1234)

# Dividing the dataset into train and test dataset
pd = sample(2,nrow(cardio),replace= TRUE,prob=c(0.8,0.2))

# Train dataset
train = cardio[pd==1,] 

# No. of record in the train dataset from cardio dataset
nrow(train)/nrow(cardio)*100

# Test dataset
test = cardio[pd==2,]

# No. of record in the test dataset from cardio dataset
nrow(test)/nrow(cardio)*100
```

*We have 80.81% and 19.19% of the records in the train and test dataset respectively from the cardio dataset.*

# **Building Neural Network**

```r
# Importing library neuralnet
library(neuralnet)

# Neural Network
nn = neuralnet(NSP ~ ., data = train)
```

# **Prediction**

```r
# Predicting neural net from test dataset
pred_nn = compute(nn, test[,-22])

# Assigning the result to p_nn
p_nn = pred_nn$net.result

# Prediction
head(p_nn, 5)

# Actual values
head(test[,22], 5)
```

```r
# Prediction vs Actual Value
cbind(head(p_nn, 15), head(test[,22],15))
```

*The prediction for 5th record is 1.013863 which is close to 1, and this is a accurate prediction. Similarly the prediction is accurat for all other records shown above. Except for 90th we the prediction is close to 2 (1.180691), but the actual value is 2.*


```r
# Prediction vs Actual Value
cbind(tail(p_nn, 15), tail(test[,22],15))
```

*We can see some **wrong** prediction in the above table. Like for 2061th the value is 1.740470 which is close to 2 but the actual calue is 1. Similary the prediction is wrong in case of 2066th, 2089th, 2091th and 2116th.*

# **Accuracy Level**

```r
# Predicting the value as categories 1, 2 or 3
# Assign 3 if the value is more than 2.5, if more than 1.5, otherwise 1.
cm = ifelse(p_nn > 2.5, 3, ifelse(p_nn > 1.5, 2, 1))

# Confusion Matrix
tab_nn=table(cm,test$NSP)

# Printing Confusion Matrix
print(tab_nn)
```

```r
# Accuracy Percentage
sum(diag(tab_nn))/sum(tab_nn)*100
```

*The network is giving 85.05% accuracy. For this high value we can consider the network, and the model is likely to give correct prediction 85.05% of the time.*
